```python
import numpy as np
import pandas as pd
import copy as cp
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
df0 = pd.read_csv("data.csv")
```

```python
df0.head()
```

```python
df0.columns
```

```python
df0 = df0.drop(['Age-adjusted Rate',
       'Standard Error Age-adjusted Rate',
       'Lower Confidence Limit for Age-adjusted rate',
       'Upper Confidence Limit for Age-adjusted Rate',
       'State Crude Rate in Range'], axis=1)
```

```python
df0.head()
```

```python
df0['Rate'] = 100000 * df0['Deaths'] / df0['Population']
```

```python
df0.head()
```

```python
df0 = df0.drop(['Crude Death Rate', 'Standard Error for Crude Rate',
       'Low Confidence Limit for Crude Rate',
       'Upper Confidence Limit for Crude Rate', 'US Crude Rate', 'US Age-adjusted Rate'], axis=1)
```

```python
df0.head()
```

```python
df0 = df0.loc[df0['Sex'] == 'Both Sexes']
df0.head()
```

```python
df0 = df0.loc[df0['Race and Hispanic Origin'] == 'All Races-All Origins']
df0.head()
```

```python
df0 = df0.loc[df0['State'] == 'United States']
df0.head()
```

```python
df0 = df0.loc[df0['Age'] != 'All Ages']
df0.head()
```

```python
df0.shape
```

```python
df0.plot.scatter(x='Population', y='Deaths', c='Rate', colormap='viridis')
```

```python
groups = df0.groupby('Age')

fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group['Population'], group['Deaths'], marker='o', linestyle='', ms=5, label=name)
ax.legend()

plt.show()
```

```python
from sklearn import preprocessing

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df0o = pd.get_dummies(df0, columns=['Year', 'Age', 'Sex', 'State', 'Race and Hispanic Origin'])
df0o.head()
```

```python
dt_mse = DecisionTreeRegressor(random_state=0, criterion='mse')
dt_mse.fit(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1), df0o['Rate'])
```

```python
dt_mse.feature_importances_
```

```python
for i in range(len(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1).columns)):
    print(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1).columns[i], "has importance", dt_mse.feature_importances_[i])
```

```python
rf_mse = RandomForestRegressor(n_estimators=50000, n_jobs=8, criterion='mse', random_state=0)
rf_mse.fit(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1), df0o['Rate'])
```

```python
for i in range(len(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1).columns)):
    print(df0o.drop(['Deaths', 'Population', 'Rate'],axis=1).columns[i], "has importance", rf_mse.feature_importances_[i])
```

## ANOVA

```python
from sklearn import svm
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X = df0o.drop(['Deaths', 'Population', 'Rate'],axis=1)
y = df0o['Rate']
```

```python
anova_filter = SelectKBest(f_regression, k=8)
clf = svm.LinearSVR
```

```python
anova_svm = make_pipeline(anova_filter, clf)
```

```python
anova_svm.fit(X,y)
```
